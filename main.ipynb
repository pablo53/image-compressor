```python
# Image compressor
```

```python
import matplotlib.image as img
import matplotlib.pyplot as plt
import numpy as np
```

```python
### Loading an image
```

```python
image = img.imread('Nadkole.jpg')
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
### B&W image:
```

```python
bwimage = np.sum(image, axis=2, dtype=np.float64) / 765.0
```

```python
bwimage.shape
```

```python
plt.imshow(bwimage, cmap='gray')
```

```python
### Computing means over axes:
```

```python
bwx = np.mean(bwimage, axis=0)
bwy = np.mean(bwimage, axis=1)
```

```python
bwx.shape
```

```python
bwy.shape
```

```python
bwxy = np.kron(bwx.reshape((1, bwx.shape[0])), bwy.reshape((bwy.shape[0], 1)))
```

```python
bwxy.shape
```

```python
plt.imshow(bwxy, cmap='gray')
```

```python
### Reducing the original image:
```

```python
bwimage1 = bwimage / bwxy
```

```python
plt.imshow(bwimage1, cmap='gray')
```
